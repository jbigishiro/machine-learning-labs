```python
names = ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Hannah", 
         "Ian", "Jack", "Katie", "Leo", "Mia", "Nina", "Oscar", "Paul", 
         "Quinn", "Rachel", "Sam", "Tina", "Uma", "Violet", "Will", 
         "Xander", "Yara", "Zane", "Charlie", "David", "Eva", "Frank", "Grace", "Hannah", 
         "Ian", "Jack", "Katie", "Leo", "Mia", "Nina", "Oscar", "Paul", 
         "Quinn", "Rachel", "Sam", "Tina", "Uma", "Violet", "Will", 
         "Xander", "Yara", "Zane"]
```

```python
grade = np.random.choice(["A", "B", "C", "D", "F"], 50)
```

```python
obedient = np.random.choice(["Y", "N"], 50)
```

```python
researchScore = np.random.randint(50, 100, 50)
```

```python
projectScore = np.random.randint(50, 100, 50)
```

```python
recommend = np.random.choice(["Y", "N"], 50)
```

```python
df = pd.DataFrame({"Name": names, "overallGrade": grade, "Obedient":obedient, "ResearchScore": researchScore, "ProjectScore": projectScore, "Recommend": recommend })
```

```python
df.columns
```

```python
X = df[['overallGrade', 'Obedient', 'ResearchScore', 'ProjectScore']]
y = df[["Recommend"]]
```

We will now use a standard scalar from scikit-learn to scale or normalize our two numeric scorebased
attributes using the following code.

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(X[['ResearchScore', 'ProjectScore']])
X[['ResearchScore', 'ProjectScore']] = ss.transform( X[['ResearchScore', 'ProjectScore']])
```

```python
X = pd.get_dummies(X, columns=['overallGrade', 'Obedient'])
```

```python
X
```

### Modeling

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X, y)
```

### Model Evaluation

```python
yhat = logreg.predict(X)
```

```python
yhat
```

```python
y
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
accuracy_score(y, yhat)
```

```python
classification_report(y, yhat)
```
